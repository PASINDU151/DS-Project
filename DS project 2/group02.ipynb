```python
from sklearn.ensemble import RandomForestRegressor

train_data = Marvellous[Marvellous['Basic Salary_0'].notnull()]
predict_data = Marvellous[Marvellous['Basic Salary_0'].isnull()]

train_data['Gender'] = train_data['Gender'].map({'Male': 0, 'Female': 1})
predict_data['Gender'] = predict_data['Gender'].map({'Male': 0, 'Female': 1})

# Convert 'Date_Joined' to datetime if it's not already
train_data['Date_Joined'] = pd.to_datetime(train_data['Date_Joined'])
predict_data['Date_Joined'] = pd.to_datetime(predict_data['Date_Joined'])

# Extract year, month, and day
train_data['Year_Joined'] = train_data['Date_Joined'].dt.year
train_data['Month_Joined'] = train_data['Date_Joined'].dt.month
train_data['Day_Joined'] = train_data['Date_Joined'].dt.day

predict_data['Year_Joined'] = predict_data['Date_Joined'].dt.year
predict_data['Month_Joined'] = predict_data['Date_Joined'].dt.month
predict_data['Day_Joined'] = predict_data['Date_Joined'].dt.day

ordinal_mapping = {'Labour': 3, 'Staff': 2, 'Management': 1}

train_data['Employment_Category'] = train_data['Employment_Category'].map(ordinal_mapping)
predict_data['Employment_Category'] = predict_data['Employment_Category'].map(ordinal_mapping)

features = ['Gender', 'Designation_ID','Year_Joined', 'Month_Joined', 'Day_Joined','Employment_Category']

# Use all other columns as features
X_train = train_data[features]
X_predict = predict_data[features]
y_train = train_data['Basic Salary_0']
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train, y_train)
```

```python
# Predict the null values
predicted_values = model.predict(X_predict)

# Fill the null values in the original dataframe
Marvellous.loc[Marvellous['Basic Salary_0'].isnull(), 'Basic Salary_0'] = predicted_values
```

```python
import numpy as np

single_count = Marvellous['Marital_Status'].value_counts().get('Single', 0)
married_count = Marvellous['Marital_Status'].value_counts().get('Married', 0)

total_count = single_count + married_count
single_proportion = single_count / total_count
married_proportion = married_count / total_count

null_count = Marvellous['Marital_Status'].isnull().sum()

new_single = int(null_count * single_proportion)

null_indices = Marvellous[Marvellous['Marital_Status'].isnull()].index

single_indices = np.random.choice(null_indices, size=new_single, replace=False)
Marvellous.loc[single_indices, 'Marital_Status'] = 'Single'

Marvellous.loc[Marvellous['Marital_Status'].isnull(), 'Marital_Status'] = 'Married'

remaining_nulls = Marvellous['Marital_Status'].isnull().sum()
print("Remaining null values in Marital_Status column:", remaining_nulls)
```

```python
%pip install kmodes
from kmodes.kmodes import KModes

features = ['Designation_ID', 'Marital_Status', 'Employment_Type', 'Year_of_Birth'] # features for clustering ... assuming these attributes are related to the year of birth

# data as the working dataset
data = Marvellous[features]
# data= data.dropna(subset=['Marital_Status'])   ... was used for testing purposes

cluster_features = ['Designation_ID', 'Marital_Status', 'Employment_Type']  # features for clustering
cl_data = data[cluster_features]

# clusters are created based on the above columns using kmodes algorithm, assuming 5 clusters
k = 5  # Number of clusters
k_modes = KModes(n_clusters=k, init='Huang', n_init=5, verbose=1)

clusters = k_modes.fit_predict(cl_data)

data['Cluster'] = clusters

cluster_group = data.groupby('Cluster')

# Calculate the mode of 'Year_of_Birth' for each cluster
cluster_modes = cluster_group['Year_of_Birth'].apply(lambda x: x.value_counts().index[1] if x.mode()[0]=="'0000'" else x.mode()[0])

print(cluster_modes)

cluster_mode_dict = cluster_modes.to_dict() #  dictionary of cluster modes
print(cluster_mode_dict)

# Replacing '0000' values in 'Year_of_Birth' with mode value based on cluster
data['Year_of_Birth'] = data.apply(lambda row: cluster_mode_dict[row['Cluster']] if row['Year_of_Birth'] == "'0000'" else row['Year_of_Birth'], axis=1)

Marvellous['Year_of_Birth'] = data['Year_of_Birth']
```

```python
def convert_date_format(date):
    if pd.isnull(date):
        return date  # Return NaN if the date is missing
    try:
        parsed_date = datetime.strptime(str(date), "%m/%d/%Y")
        formatted_date = parsed_date.strftime("%Y-%m-%d")
        return formatted_date
    except ValueError:
        return str(date)  # Return the original value if parsing fails

# Assuming Marvellous is the name of your DataFrame
Marvellous['Date_Joined'] = Marvellous['Date_Joined'].apply(lambda x: convert_date_format(x))
Marvellous['Date_Resigned'] = Marvellous['Date_Resigned'].apply(lambda x: convert_date_format(x))
Marvellous['Inactive_Date'] = Marvellous['Inactive_Date'].apply(lambda x: convert_date_format(x))
```

Hypothesis 02

```python
import pandas as pd
from scipy.stats import chi2_contingency

 
Marvellous['Date_Resigned'] = Marvellous['Date_Resigned'].replace('0000-00-00', pd.NaT)

# Convert 'Date_Resigned' column to datetime
Marvellous['Date_Resigned'] = pd.to_datetime(Marvellous['Date_Resigned'], errors='coerce')

# Exclude rows with NaT (invalid) values in 'Date_Resigned' column
resigned_employees = Marvellous.dropna(subset=['Date_Resigned'])

# Create a contingency table between 'Employment_Category' and 'Date_Resigned'
contingency_table = pd.crosstab(resigned_employees['Employment_Category'], 
                                resigned_employees['Date_Resigned'].dt.year)

# Perform chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Define significance levels
alpha_levels = [0.05, 0.1, 0.2]
# Print Chi-square statistic and p-value
print("Chi-square statistic:", chi2)
print("p-value:", p_value)

# Print hypothesis testing results for different significance levels
print("Hypothesis testing result:")
for alpha in alpha_levels:
    if p_value < alpha:
        print(f"H0 is rejected at {alpha*100:.1f}% significance level.")
    else:
        print(f"H0 is not rejected at {alpha*100:.1f}% significance level.")

# Print conclusion
if p_value < min(alpha_levels):
    print("\nIt seems there is a significant impact of employment category on employee attrition.")
else:
    print("\nThere is not enough evidence to conclude a significant impact of employment category on employee attrition.")
```
